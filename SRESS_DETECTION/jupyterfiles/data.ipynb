```python
import csv
import os

# Define the input CSV file path
input_csv_path = 'data.csv'

# Create a directory to store the individual CSV files
output_dir = 'Test_data'
# os.makedirs(output_dir, exist_ok=True)

# # Read the input CSV file
with open(input_csv_path, 'r') as csvfile:
    # Initialize a CSV reader
    csvreader = csv.reader(csvfile)
    
    # Skip the header row if it exists
    header = next(csvreader, None)
    
    # Loop through the rows
    for i, row in enumerate(csvreader):
        # Create a unique file name based on the row index
        output_csv_path = os.path.join(output_dir, f'Record_{i}.csv')
        
        # Write the row to the individual CSV file
        with open(output_csv_path, 'w', newline='') as output_csvfile:
            csvwriter = csv.writer(output_csvfile)
            csvwriter.writerow(header)  # Write the header
            csvwriter.writerow(row)     # Write the row data
```
